```python
SKU_w_orderno = pd.merge(SKU_master,order_report, on= 'SKU')
```

```python
SKU_w_orderno['Weight by Qty'] = SKU_w_orderno['Weight (g)'] * SKU_w_orderno['Order Qty']
SKU_w_orderno
```

```python
temp = SKU_w_orderno.groupby('ExternOrderNo')['Weight by Qty'].sum()
temp = temp.reset_index()
```

```python
temp.rename(columns = {'Weight by Qty':'Total_weight'}, inplace = True)
```

```python
temp
```

```python
import numpy as np
```

```python
temp['D_500'] = temp['Total_weight'] /500
temp['D_500'] = np.ceil(temp['D_500'])
```

```python
temp['weight_slab_x'] = temp['D_500'] * 0.5
temp
```

```python
temp.rename(columns = {'ExternOrderNo':'Order ID'}, inplace = True)
temp
```

```python
temp
```

```python
temp1 = pd.merge(temp,CC_invoice, on = 'Order ID')
temp1
```

```python
temp1.drop(['Warehouse Pincode'], axis =1 , inplace = True)
```

```python
temp1.rename(columns = {'Charged Weight':'Total_weight_cc'}, inplace = True)
```

```python
temp1.rename(columns = {'Zone':'Zone_cc'}, inplace = True)
```

```python
temp1.rename(columns = {'Billing Amount (Rs.)':'Amount_charge_cc'}, inplace = True)
```

```python
temp1
```

```python
temp1['Total_weight_cc'] = temp1['Total_weight_cc'].apply(lambda x:x*1000)
```

```python
temp1
```

```python
temp1['D_500'] = temp1['Total_weight_cc'] / 500
temp1['D_500'] = np.ceil(temp1['D_500'])
temp1['weight_slab_cc'] = temp1['D_500'] * 0.5
temp1
```

```python
temp1.drop(['D_500'], axis = 1 , inplace = True)
```

```python
temp1
```

```python
temp2 = pd.merge(temp1 ,pincode , on = 'Customer Pincode')
```

```python
temp2
```

```python
temp2.drop(['Warehouse Pincode'], axis = 1 , inplace = True)
temp2.drop(['Customer Pincode'], axis = 1 , inplace = True)
temp2
```

```python
temp2.isnull().sum()
```

```python
temp2.info()
```

```python
list_zone = temp2['Zone_x'].unique()
list_zone
```

```python
final_df = pd.DataFrame()
final_df
```

```python
value_pair = {'a':29.5,'b':33,'c':40.1,'d':45.4,'e':56.6}
```

```python
additional_charges = {'a':23.6,'b':28.3,'c':38.9,'d':44.8,'e':55.5}
```

```python
rto_fixed_charges = {'a':13.6,'b':20.5,'c':31.9,'d':41.3,'e':50.7}
```

```python
rto_additional_charges = {'a':23.6,'b':28.3,'c':38.9,'d':44.8,'e':55.5}
```

```python
for i in list_zone:
    temp3 = temp2[temp2['Zone_x'] == i]
    temp3['fixed'] = value_pair[i]
    temp3['additional'] = np.where(temp3['weight_slab_x'] *2 > 1, ((temp3['weight_slab_x']*2)-1) * additional_charges[i],0)
    temp3['rto-fixed_charges'] = np.where(temp3['Type of Shipment'].str.contains('RTO', case = False), rto_fixed_charges[i],0)
    temp3['rto_additional_charges'] = np.where((temp3['weight_slab_x'] *2 >1) & (temp3['Type of Shipment'].str.contains('Forward and RTO charges', case = False)), ((temp3['weight_slab_x']*2)-1) * rto_additional_charges[i],0)
    final_df = pd.concat([final_df, temp3])
```

```python
final_df.rename(columns = {'rto-fixed_charges':'rto_fixed_charges'}, inplace = True)
```

```python
final_df['expected_charges_X'] = final_df['fixed'] + final_df['additional'] + final_df['rto_fixed_charges'] + final_df['rto_additional_charges']
final_df
```

```python
final_df['difference_x_cc'] = round(final_df['expected_charges_X'] - final_df['Amount_charge_cc'],2)
final_df
```

```python
final_df = final_df.drop(columns = ['fixed','additional','rto_fixed_charges','rto_additional_charges'])
```

```python
DF = final_df[['Order ID','AWB Code','Total_weight','weight_slab_x','Total_weight_cc','weight_slab_cc','Zone_x','Zone_cc','expected_charges_X','Amount_charge_cc','difference_x_cc']]
DF
```

```python
DF.to_csv('DF.csv', index=False)
```
